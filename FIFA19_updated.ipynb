## Linear Regression 

### Predicting the Potential of a player based on his strength

```python
df_new.head()
```

### Strength vs Potential

```python
plt.figure(figsize=[10,5])
sb.jointplot(x=df_new.Strength,y=df_new.Potential,kind='hex',color='green')
plt.show()
```

```python
X=df_new['Strength']
```

```python
y=df_new['Potential']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn import metrics
```

```python
Xtrain, Xtest, ytrain,ytest=train_test_split(X,y,test_size=0.3, random_state=80)
```

```python
regressor=LinearRegression()
```

```python
type(Xtrain)
```

```python
Xtrain=np.array(Xtrain)
```

```python
type(Xtrain)
```

```python
ytrain=np.array(ytrain)
```

```python
type(ytrain)
```

```python
Xtrain=Xtrain.reshape(-1,1)
ytrain=ytrain.reshape(-1,1)
```

```python
Xtest=np.array(Xtest)
```

```python
Xtest=Xtest.reshape(-1,1)
```

```python
regressor.fit(Xtrain,ytrain)
```

```python
ypred=regressor.predict(Xtest)
```

```python
print("Mean Squared Error: "+str(metrics.mean_squared_error(ytest,ypred)))
```

## Plotting the training data

```python
plt.scatter(Xtrain,ytrain,color='blue',alpha=0.3)
plt.xlabel('Strength of player')
plt.ylabel('Potential')
plt.title('Linear regression')
plt.plot(Xtrain,regressor.predict(Xtrain),color='red')
plt.show()
```

## Plotting the test data

```python
plt.scatter(Xtest,ytest,color='blue',alpha=0.3)
plt.xlabel('Strength of player')
plt.ylabel('Potential')
plt.title('Linear regression')
plt.plot(Xtest,ypred,color='red')
plt.show()
```

```python
print("Intercept: "+str(regressor.intercept_))
```

```python
print("Coefficient: "+str(regressor.coef_))
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
X=df_new['Strength']
```

```python
y=df_new['Potential']
```

```python
Xtrain, Xtest, ytrain,ytest=train_test_split(X,y,test_size=0.3, random_state=80)
```

```python
regressor2=LinearRegression()
```

```python
Xtrain=np.array(Xtrain)
ytrain=np.array(ytrain)
```

```python
Xtrain=Xtrain.reshape(-1,1)
ytrain=ytrain.reshape(-1,1)
```

```python
Xtest=np.array(Xtest)
```

```python
Xtest=Xtest.reshape(-1,1)
```

```python
poly_regression=PolynomialFeatures(degree=3)
```

```python
X=np.array(X)
```

```python
X=X.reshape(-1,1)
```

```python
polynomial_X=poly_regression.fit_transform(X)
```

```python
polynomial_X
```

```python
x_polytrain,x_polytest,y_train,y_test=train_test_split(polynomial_X,y,test_size=0.2, random_state=42)
```

```python
regressor2.fit(x_polytrain,y_train)
```

```python
ypred2=regressor2.predict(x_polytest)
```

```python
plt.scatter(Xtest,ytest,color='yellow')
plt.xlabel("Strength")
plt.ylabel("Potential")
plt.title("Polynomial Regression")
plt.plot(Xtrain,regressor2.predict(poly_regression.fit_transform(Xtrain)),color='black')
plt.show()
```

```python
print("Intercept: "+str(regressor2.intercept_))
```

```python
print("Coefficient: "+str(regressor2.coef_))
```

```python
print("Mean Squared Error: "+str(metrics.mean_squared_error(y_test,ypred2)))
```

### As Mean squared error of Polynomial Regression is less than that of Linear Regression, therefore Polynomial Regression provides more accurate results.
